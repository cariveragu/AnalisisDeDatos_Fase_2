```python
#Importar librerias
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn import linear_model

import seaborn as sns
```

```python
#Guardar en variable los datos locales
datos = pd.read_csv("C:/Users/1/OneDrive - Universidad Nacional Abierta y a Distancia/Periodo 2023 II 16-04 (1394)/1 - ANALISIS DE DATOS - (202016908)/Etapa 2/RegresionLineal/data.csv", sep = ",")
```

```python
#Revisar Variable
datos
```

```python
#Grafica de dispersión
datos.plot.scatter(x= "metro", y= "precio")
plt.show()
```

```python
#Variable que contendra la regresión lineal
regresion = linear_model.LinearRegression()
```

```python
#Agregar los datos en un vector o array
metros = datos["metro"].values.reshape((-1,1))
```

```python
#crear el modelo
modelo = regresion.fit(metros, datos["precio"])

#Imprimir la intersección
print("Intersección (b)", modelo.intercept_)

#imprimir la pendiente
print("Pendiente (m)", modelo.coef_)
```

```python
#Realizar predición
entrada = [[18],[22],[27]]
prediccion = modelo.predict(entrada)
print(prediccion)
```

```python
#creamos el gráfico de la regresión lineal
datos.plot.scatter(x = "metro", y = "precio", label='Datos')
plt.scatter(entrada, prediccion, color='red')
plt.plot(entrada, prediccion, color='black', label='Regresión lineal')
plt.xlabel('Metros')
plt.ylabel('Precios')
plt.legend()
plt.show()
```
