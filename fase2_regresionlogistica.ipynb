```python
# pruebas de parametro
w = 0.09
b = -3.6
```

```python
x = np.linspace(0,datos['age'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))

# grafica de la recta
datos.plot.scatter(x='age',y='glucose')
#plt.plot(x, y, '-r')
plt.plot(x, y, color = 'black')
plt.ylim(0,datos['glucose'].max()*1.1)
plt.scatter(x, y, color='#A9E2F3')
# plt.grid()
plt.xlabel('Edad')
plt.ylabel('Glucosa')
plt.show()
```
